```python
import pandas as pd
import numpy as np
from scipy.io import savemat
import os
import pickle
```

```python
#src = r"\\10.99.68.178\andreex\data\monkey fetus\gestational 40\hovernet_out\df"
#dfs = sorted([_ for _ in os.listdir(src) if _.endswith('pkl')])

#src = r'\\10.162.80.16\Andre_expansion\data\monkey_fetus\Stardist\StarDist_12_25_23\stardist_feature_df_pickles'
# src = r'\\10.162.80.16\Andre_expansion\data\monkey_fetus\Stardist\StarDist_12_25_23\dfs_fixed_centroids_3_14'
pth_WSI_ndpi = r'\\10.162.80.16\Andre\data\monkey fetus\bissected_monkey_GS55'
pth_segmentation_json = os.path.join(pth_WSI_ndpi,'StarDist_12_12_23', 'json')
pth_pixel_res = os.path.join(pth_WSI_ndpi,'segmentation_analysis', 'pix_res_info')
src = os.path.join(pth_WSI_ndpi,'StarDist_12_12_23', 'stardist_feature_df_pickles')
outpth=os.path.join(src,'mat')
os.makedirs(outpth, exist_ok=True)

dfs = sorted([_ for _ in os.listdir(src) if _.endswith('.pkl')])
```

```python
print(dfs[0:1])
```

```python
for dfnm in dfs:
    print("Saving: {}".format(dfnm))
    dst = os.path.join(*[outpth,dfnm.replace('pkl','mat')])
    if os.path.exists(dst):
        print("mat already exists, skipping the file ID {}".format(dfnm))
        continue
        
    with open(os.path.join(src, dfnm), 'rb') as f:
        df = pickle.load(f)

    # df = pd.read_pickle(os.path.join(src,dfnm))
    # df = df[['Centroid_x', 'Centroid_y']]
    col_names = df.columns.tolist()
    df = [_ for _ in df.to_numpy()]
    df = np.array(df)
    print(dst)
    
    savemat(dst, {'features':df, 'feature_names':col_names})
```
